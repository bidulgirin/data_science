여러분은 왜 개 사료가 예상보다 빨리 상하게 되는지를 파악하기위해 개 사료 회사에 고용되었습니다. 안 좋은 소식은 이 회사는 한동안 장비를 업그레이드 하지 않았습니다. 즉, 이 회사에서 사용하는 5가지 방부제의 사용 용량이 중구난방이라는 뜻입니다. 하지만 그 중에서 가장 많은 영향을 끼치는 것이 무엇일까요?<br>
이 회사에서는 먼저 4가지 방부제(A, B, C, D)가 함유된 방부제 묶음을 섞은 뒤 '필러'라는 화학물질로 마무리합니다. 식품과학자들은 A, B, C, D 방부제 중 하나가 문제를 일으키고 있다고 생각하지만, 그게 어떤 것인지 명확하게 하기위해 여러분이 고용됐습니다. "랜덤포레스트"로 어떤 방부제가 사료를 빨리 상하게 하는지 찾아보세요!

- Pres_A : 혼합물의 A비율
- Pres_B : 혼합물의 B비율
- Pres_C : 혼합물의 C비율
- Pres_D : 혼합물의 D비율
- Spoiled : 개 사료가 상했는지 여부를 나타내는 레이블.

**시나리오에서 요구하는 바를 고려하여, 머신러닝을 이용해서 문제를 풀어보세요.**

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.model_selection import train_test_split, cross_validate
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier, plot_tree #트리기반의 모델들
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor # 다중공선성체크해봄
```

```python
# 예상보다 빨리 상하게 되는지를 파악

# 한동안 장비를 업그레이드 하지 않았습니다
# 5가지 방부제의 사용 용량이 중구난방
# 가장 많은 영향을 끼치는 것

# "랜덤포레스트"로 어떤 방부제가 사료를 빨리 상하게 하는지

# 근거들을 나열 

# 필러라는게 꼬롬하이 그래

# EDA 시각화 진행
```

# 일단 독립변수들의 관계를 알아보자

```python
# 독립, 종속 담아~~
x = df.drop("Spoiled", axis = 1 ) 
y = df["Spoiled"]
```

```python
x = sm.add_constant(x) # 절편 추가
model = sm.OLS(y,x)
model = model.fit()
model.summary()
```

```python
# 나눠 나눠
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size = 0.2, random_state=666
)
```

```python
# 의사결정나무니까 함 시각화 자료 그려보자
dt = DecisionTreeClassifier(random_state=666)
dt.fit(x_train, y_train)
```

```python
plt.figure()
plot_tree(dt, max_depth = 1, filled = True, feature_names = ["A", "B","C","D"])
plt.show()
```

```python
# 특성 중요도 찍어보자
print(dt.feature_importances_)
# 지니계수 왜 이리 낮음?
```

- 특성 중요도는 C, D, B, A 이런 순서가 나온다

```python
# 특성 중요도 라는게 가장 유용한지라고 나타내는 지표라고 하는데 
# 상관관계가 높다는 뜻일까?
```

# 독립변수끼리 다중공선성의 문제가 있는가?

```python
vif = pd.DataFrame()
vif["VIF_Factor"] = [variance_inflation_factor(x.values, i) for i in range(x.shape[1])]
vif["feature"] = x.columns
vif.round(1)
```

- VIF 가 15가 넘지 않으므로 딱히 독립변수끼리 방해를 하고 있진 않음
