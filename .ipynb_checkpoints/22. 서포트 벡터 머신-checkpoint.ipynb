# 서포트 벡터 머신

<img src="./image/svm.png">

-  범주를 나눠줄 수 있는 최적의 구분선 (결정경계선)을 찾아내서 고나측치의 범주를 예측하는 모델
-  로지스틱 회귀나 판별분서에 비해 비선형 데이터에서 높은 정확도를 보이며 다른 모델들 보다 과적합되는 경향이 적음
-  SVM 결정경계선의 양쪽이 빈 공간 마진을 최대화하도록 만들어졌음

- 마진과 맞닿아서 결정경계선의 위치와 각도를 정해줄 수 있는 기준이 되는 관측치를 서포트 벡터라고 함
- 결정 경계선을 지지해주고 있기 때문에
- svm 은 서포트 벡터만으로 범주의 구분 기준인 결정경계선을 정하기 때문에 학습효율이 높음
- 마진은 결정 경계선과 서포트 벡터와의 거리를 의미
- 기계학습에서 거리를 통해 분류나 회귀 모델을 만드는 경우에는 데이터 정규화나 표준화를 해주는 것을 추천
- 변수마다 단위가 다르게 되면 모델 성능이 매우떨어지기때문
- 독립변수가 2개라면 2차원사으이 결정경계선을 표현 할 수 있음
    - 독립변수가 3개로 늘어나면 평면이 됨
    - 그 이상으로 차원이 증가하면 그림으로 표혀할수없는 초평면이 결정 경계선이됨
    - 따라서 svm 은 mmh 최대 마진으로 찾아 범주를 분리하는 분류방법
- 하지만 대부분의 데이터는 이상치가 있을 수 있음
- 결정 경계선을 만들 때 이상치를 허용하지 않는 경우에는 과적합 문제가 발생할 가능성이 있음
- 따라서 어느정도 이상치를 허용해주도록 하여 두 범주를 정확하게 나누지는 않지만 마진을 최대화하여 과적합 방지
    - 소프트 마진이라캄
    - 이상치 허용 안하면 하드 마진

<img src="./image/svm_soft_hard.png">

- 소프트 마진과 하드마진을 조정해주는 매개변수로는 C와 gamma 가 사용됨
- C값을 낮게 설정하면 이상치들이 있을 가능성을 높게 잡아서 마진 오류를 더 많이 허용함
- 마진 오류란 ? 결정 경계와 마진사이, 또는 마진 내부에 위치하는 훈련 데이터 포인트
- 너무 낮추면 ? 모델이 단순~ 일반화 성능 높아짐 : 과소적합 문제가 있을 수 있음
- 너무 높이면 ? 과대적합

- 이는 모델이 훈려 넫이터에 더 정확히 맞추려고 하지만 동시에 과대적합될 위험 커짐
- 따라서 훈련 데이터에서는 잘 맞지만 새로운 데이터에는 잘 일반화되지 않을 수 있음

- gamma 값은 관측치가 영향력을 행사하는 거리를 조정해 주는것

- 값이 클수록 영향력의 거리는 짧아지고, 값이 작을수록 영향력의거리는 길어짐

- 커질수록 각각의 관측치에 대한 결정경계선이 범위 작아짐

- gamma 값이 커지면 여러 개의 결정경계선 생기게됨

<img src="./image/svm_gamma.png">

- 위와 같은 데이터 분포에서는 아무리 소프트마진을 높게 설정하더라도 결정경계선을 만들기 힘들고 , 만든다고 하더라도 성능이 떨어지게 됨
    - 그런 경우 사용하는 것이 커널 기법

- 커널 기법으 기존의 데이터를 고차원으로 확장하여 새로운 결정경계선을 만들어내느 방법
    - 커널 기법에는 Polynomial 커널, sigmoid 커널, 가우시안 RBF 커널 등 다양한 종류가 있음
      
- sVM 모델을 만들 때는 Gamma값을 잘 조정해 가면서 모델의 복잡도를 적정수준으로 맞춰 줘야함
- 정확도, 검증 데이터셋의 오분류율 증감 확인해가면서 과대, 과소적합 사의 최적모델 만들기

```python
import pandas as pd
import mglearn
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC # 서포터 벡터머신 (분류할꺼야)
# SVR 은 뭐징 Regresser 회귀네...~~~
from sklearn.decomposition import PCA
import warnings
warnings.filterwarnings("ignore")
```

```python
# mglearn 결정경계선 시각화
```

```python
df = pd.read_excel("./data/Raisin_Dataset.xlsx")
```

```python
df.head()
```

```python
# 독립변수 스케일링
x = df.drop(["Area","Class"], axis = 1)
y = df["Class"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x,
    y,
    test_size = 0.3,
     stratify= y,
    random_state= 33
)
```

```python
mm = MinMaxScaler()
scaled_x_train = mm.fit_transform(x_train)
scaled_x_test = mm.transform(x_test)
```

```python
print(scaled_x_train.shape,scaled_x_test.shape, )
```

```python
# 기본 SVM 모델 성능 확인
svm = SVC()
svm.fit(scaled_x_train, y_train)

print(svm.score(scaled_x_train, y_train))
print(svm.score(scaled_x_test, y_test))
```

```python
# 6차원 친구를 차원 축소 할끄니까
pca = PCA(n_components=2)
pca_data = pca.fit_transform(scaled_x_train)
```

```python
pca_data
```

```python
# SVC C 값 1~ 30 에 따른 모델 성능 확인

scores = []

for c_point in range(1, 31):
    svc = SVC(C = c_point)
    svc.fit(scaled_x_train, y_train)
    train_score = svc.score(scaled_x_train, y_train)
    test_score = svc.score(scaled_x_test, y_test)

    print(f"SVM : C : {c_point} {train_score:.4f} , {test_score:.4f}")

    scores.append([train_score, test_score ])
```

```python
fig, axes = plt.subplots(1,3, figsize = (15, 5))

# C값이 1, 15, 30 에 따른 결정경계 시각화

for c_point, ax in zip([1,15,30] , axes):
    c_svc_model = SVC(gamma = 0.5, C = c_point)
    c_svc_model.fit(pca_data, y_train)

    mglearn.plots.plot_2d_separator(c_svc_model, pca_data, fill = True, eps = 0.5 , ax = ax, alpha = 0.2)
    mglearn.discrete_scatter(
        pca_data[:, 0 ], pca_data[:, 1], y_train, markeredgewidth = 0.2, c = ["b","r"], s = 3, ax = ax    )


    ax.set_xlabel("component 1")
    ax.set_ylabel("component 2")
    ax.set_title(f"C = {c_point}")
plt.show()
```

```python
# SVM gamma 값 1~ 30 성능 확인

scores = []

for gamaa_point in [0.1, 0.5, 1,5,10,20,30]:

    c_svc_model = SVC(gamma = 0.5, C = c_point)
    c_svc_model.fit(scaled_x_train, y_train)

    train_score = svc.score(scaled_x_train, y_train)
    test_score = svc.score(scaled_x_test, y_test)


    print(f"gamma : {gamaa_point}, train_score: {train_score:.4f}, test_score:{test_score:.4f}")

    scores.append([train_score, test_score ])
    
```

```python
fig, axes = plt.subplots(1,3, figsize = (15, 5))

# gamma 값이 0.1,10,30

for gamma, ax in zip([0.1,10,30] , axes):
    svc_model = SVC(gamma = gamma, C = 10)
    svc_model.fit(pca_data, y_train)

    mglearn.plots.plot_2d_separator(svc_model, pca_data, fill = True, eps = 0.5 , ax = ax, alpha = 0.2)
    mglearn.discrete_scatter(
        pca_data[:, 0 ], pca_data[:, 1], y_train, markeredgewidth = 0.2, c = ["b","r"], s = 3, ax = ax    )


    ax.set_xlabel("component 1")
    ax.set_ylabel("component 2")
    ax.set_title(f"gamma = {gamma}")
plt.show()
```
