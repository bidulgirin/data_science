## 모델 최적화

```python
# 독립변수를 잘 선택하려면 **기준** 이 필요하다
```

```python
# statsmodel 라이브러리로 회귀 분석
```

# 상수항 추가(절편추가하려고)

```python
x = sm.add_constant(x)
```

```python
x.head()
```

```python
# 선형회귀 모델 (오른쪽에서 왼쪽으로 가는 논리)
```

```python
model = sm.OLS(y,x)
```

```python
model = model.fit()
```

```python
model.summary()
```

- p_value 가 낮게 나오면 걍 독립변수에서 없애도 된다~ 라고 판단하면 된다
- F- statistic : 회귀식의 유의성 검정에 사용되는 갑승로, Prob(F-statistic)과 함께 해석해야함
- prob(F-statistic) : F통계량에 대한 p-value .
- P>|t| : 각 피쳐가 얼마나유의미한지

```python
# 계수 시각화
# 회귀 계수를 시리즈로 변환
coefs = model.params.tolist() # model.params => coef => 기울기네~
coefs_se = pd.Series(coefs)
```

```python
coefs_se.head()
```

```python
# 변수명을 리스트로 변환
x_labels = model.params.index.tolist()
```

```python
ax = coefs_se.plot(kind = "bar")
ax.set_title("")
ax.set_xlabel("feature")
ax.set_ylabel("coef")
ax.set_xticklabels(x_labels)
plt.show()
```

```python
# 피처들의 상관관계 시각화
corr = scaled_pich_df[scale_col].corr()

plt.figure(figsize = (15,15))

sns.set(font_scale = 1.5, font="Malgun Gothic")
hm = sns.heatmap(
    corr.values,
    cbar = True,
    annot = True,
    square = True,
    fmt = ".2f",
    annot_kws = {"size":15},
    yticklabels = scale_col,
    xticklabels = scale_col,
)

plt.tight_layout()
plt.show()
```

- 너무 강한상관관계를 가진 독립변수는 예측에 악영향을 끼칠수있음
- 근데 그렇다고 바로 삭제하는 것은 성급함

- 독립변수를 선택하는 기준을 알아보고자 했을때 상관관계가 높은게 무조건 좋은것도아니고
- 낮은것도 무조건 좋은건 아님
- 기본 야구 도메인지식 필요할 듯

- 몇몇 피처 쌍에서 높은 연관성을 발견할 수 있음

- 회귀 분석은 피처 간의 독립성을 전제로 하는 분석이기 때문에 올바른 회귀 분석을 위해서는 연관성이 높은 피처쌍을 제거
    - 다중공선성
        - 변수 간 상관 관계가 높아 분석에 부정적인 영향을 미치는 것
        - 모델 성능을 위해 어떤 피처를 제거하는 것이 옳은 판단일지 기준 제시
        - 다중 공선성은 분산팽창요인(Variance Inflation Factor(VIF)) 이라는 계수로 평가
        - 일반적으로 VIF 가 15를 넘으면 다중공선성 문제 발생 판단

```python
# 피처마다 VIF 계수출력

vif = pd.DataFrame()
vif["VIF_Factor"] = [variance_inflation_factor(x.values, i) for i in range(x.shape[1])]
vif["feature"] = x.columns
vif.round(1)
```

- 집단 변수 학살 안됨!

#### 변수제거 1

- 제거된 변수 : p_value가 높은 변수를 제거하겠다 =>>>>>>>>>홈런/9
- 무의미한 변수가 다중공선성보다 더 심각한 문제를 유발하기 때문
- 최종적으로는 P값이 높은 변수, VIF가 높은 변수 모두 제거

```python
new_x = x.drop("홈런/9", axis = 1)
```

```python
new_x
```

```python
model = sm.OLS(y, new_x)
model = model.fit()
model.summary()
```

- AIC와 BIC 결정계수를 보면된다

```python
# 피처마다 VIF 계수출력

vif = pd.DataFrame()
vif["VIF_Factor"] = [variance_inflation_factor(new_x.values, i) for i in range(new_x.shape[1])]
vif["feature"] = new_x.columns
vif.round(1)
```

- 홈런하나 빠졌다고 VIF 가 많이 줄었네요잉~~
