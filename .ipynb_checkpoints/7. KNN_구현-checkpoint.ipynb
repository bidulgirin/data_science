```python
import pandas as pd
import numpy as np
from collections import Counter
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_excel("./data/Raisin_Dataset.xlsx")
```

```python
df.head()
```

```python
x = df.drop(["Area","Class"], axis = 1)

y = df["Class"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, stratify = y, random_state= 31)
```

```python
mm = MinMaxScaler()

scaled_train = mm.fit_transform(x_train)
scaled_test = mm.fit_transform(x_test)
```

```python
# 최근접 이웃의 수
k = 5
```

```python
# 하나의 데이터와 모든 훈련 데이터와의 거리
dist = np.linalg.norm(scaled_train - scaled_test[0], axis=1) # 모든 행들간에 거리를 구하고 싶기때문에~~~
```

```python
# KNN 클래스 정의
class KNN:
    def __init__(self, k:int = 5):
        '''
            KNN 분류기 초기화
            
            Parameters:
            k : 최근접 이웃의 수
        '''      

        self.k = k

    def fit(self, x:np.ndarray, y:np.ndarray):
        '''
            학습 데이터와 레이블을 저장
            
            Parmeters :
            x : 학습데이터, shape -> (num_samples, num_features)
            y : 레이블, shape -> (num_samples, )
        '''
        self.x_train = x
        self.y_train = y
                                    # 반환값을 입력하는거임~~~~~~~~~
    def predict(self, x:np.ndarray) -> np.ndarray:
        '''
        새로운 데이터 포인트에 대한 예측을 수행

        Parameters:
        x: 테스트 데이터 , shape -> (num_samples, num_features)

        Returns:
        예측된 레이블, shape -> (num_test_samples,)
        '''

        predictions = []

        for i in x:
            # 모든 학습 데이터와의 거리 계산 
            distances = np.linalg.norm(self.x_train - i , axis = 1)

            # 가장 가까운 k개의 인덱스
            k_indexes = distances.argsort()[:self.k]

            # 최빈값의 레이블을 예측~~~~~~!~!~!~!!~!!~!!~!~!~!~!
            k_nearest_labels = self.y_train[k_indexes]
            most_common = Counter(k_nearest_labels).most_common(1)
            predictions.append(most_common[0][0])

        return np.array(predictions)

    
```

# 거리 가중치, 측정 방식 추가

```python
# KNN 클래스 거리 가중치, 측정 방식 추가
class KNN:
    def __init__(self, k:int = 5, weighted:bool = False, metric:str = "euclidean"):
        '''
            KNN 분류기 초기화
            
            Parameters:
            k : 최근접 이웃의 수
            weighted : 
            metric : 
            
        '''      

        self.k = k
        self.weighted = weighted
        self.metric = metric

    
    def distance(self, x1, x2):
        if self.metric == "euclidean":
            return np.sqrt(np.sum((x1 - x2)**2), axis = 1)

        elif self.metric == "manhattan":
            return np.sum(np.abs(x1 - x2), axis = 1 )

        else:
            ValueError("지원 ㄴㄴ")

    
    def fit(self, x:np.ndarray, y:np.ndarray):
        '''
            학습 데이터와 레이블을 저장
            
            Parmeters :
            x : 학습데이터, shape -> (num_samples, num_features)
            y : 레이블, shape -> (num_samples, )
        '''
        self.x_train = x
        self.y_train = y
                                    # 반환값을 어떻게 내놓을지 입력하는거임 ~~~~~~~~~
    def predict(self, x:np.ndarray) -> np.ndarray:
        '''
        새로운 데이터 포인트에 대한 예측을 수행

        Parameters:
        x: 테스트 데이터 , shape -> (num_samples, num_features)

        Returns:
        예측된 레이블, shape -> (num_test_samples,)
        '''

        predictions = []

        for i in x:
            # 모든 학습 데이터와의 거리 계산 
            # distances = np.linalg.norm(self.x_train - i , axis = 1)
            distances = self.distance(self.x_train, i)

            # 가장 가까운 k개의 인덱스
            k_indexes = distances.argsort()[:self.k]

            # 최빈값의 레이블을 예측~~~~~~!~!~!~!!~!!~!!~!~!~!~!
            k_nearest_labels = self.y_train[k_indexes]



            # 가까운 거리를 크게 먼값을 작게 만들려면 역수를 만들되 0이나오면 곤란하니까 + 1e-5= 0.0000000000000...1 해줌
            if self.weighted: 
                weights = 1 / (distances[k_indexes] + 1e-5)
            
                label_weights = dict()
            
                for label, weight in zip(k_nearest_labels, weights):
                    label_weights[label] = label_weights.get(label, 0) + weight
                
                # 가장 가중치를 가진 레이블 선택
                predicted_label = max(label_weights.items(), key=lambda item: item[1])[0]
            else:
                # 거리 가중치 미사용
                most_common = Counter(k_nearest_labels).most_common(1)
                # predictions.append(most_common[0][0])
                predicted_label = most_common[0][0]
        
        
        predictions.append(predicted_label)
        return np.array(predictions)
```

```python
knn = KNN(k=5, weighted = True, metric = "manhattan")
```

```python
knn.fit(scaled_train, y_train.values)
```

```python
knn.predict(scaled_test[[0]])
```
