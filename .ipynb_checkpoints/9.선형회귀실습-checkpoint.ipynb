```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression # 선형모델들중 선형회귀모델을 불러오기
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error # 제곱해서 평균낸 잔차
from scipy import stats
```

```python
df = pd.read_csv("./data/auto-mpg.csv", header = None, names = ["mpg", "cylinders","displacement","horsepower","weight","acceleration","model_year","origin","name"])
```

```python
df.head()
```

종속변수 : mpg

```python
df.dtypes
```

```python
# 종속변수를 mpg 라고 치고 
# 독립변수는 cylinders horsepower weight
```

```python
# 전처리해야함
# horsepower 를 평균값으로 대체하겠다
```

```python
df.isna().sum()
```

# 판다스를 못해서 슬픈 사람

```python
df["horsepower"] = df["horsepower"].replace("?", np.nan)
df = df.dropna(subset = ["horsepower"], axis=0 )
```

```python
df["horsepower"] = df["horsepower"].astype(float)
```

```python
df = df[["mpg","cylinders","horsepower","weight"]]
```

```python
df.head()
```

```python
x = df[["cylinders","horsepower","weight"]] # df.drop("mpg", axis=1) 이렇게 넣어도 되는구나
y = df["mpg"]
```

```python
# 비율 나누기 
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x, 
    y,
    test_size = 0.25,
    random_state = 666
)
```

```python
# 모델 생성하기 
```

```python
lr = LinearRegression()
```

```python
# 먹이기
```

```python
lr.fit(x_train, y_train)
```

```python
# 훈련 데이터와 회귀선 시각화

plt.figure()

plt.scatter(x_train["cylinders"], y_train)
plt.xlabel("horsepower")
plt.ylabel("mpg")
plt.show()
```

```python
# 훈련 데이터와 회귀선 시각화

plt.figure()

plt.scatter(x_train["horsepower"], y_train)

plt.xlabel("horsepower")
plt.ylabel("mpg")
plt.show()
```

- 음의 상관관계가 있어 보임

```python
# 훈련 데이터와 회귀선 시각화

plt.figure()

plt.scatter(x_train["weight"], y_train)

plt.xlabel("weight")
plt.ylabel("mpg")
plt.show()
```

- weight 와 mpg 도 음의 상관관계가 있어보임

# 회귀 분석 모형

```python
r_square = lr.score(x_test, y_test)
```

```python
r_square # 회귀니까 score에서 결정계수가 나오겠져~~~(정확도 개념없음)
```

```python
print(lr.coef_, lr.intercept_)
```

회귀식 = -0.3 * 실린더 + -0.04 * 마력 + -0.005 * 무게 + 46.086970574971346

```python
y_pred = lr.predict(x_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
```

```python
mse
```

```python
# 제곱의 평균이기 때문에 제곱근 씌어줘야함
```

```python
np.sqrt(mse)
```

- 평균적으로 3~4의 오차가 있음!!!!!

```python
pd.DataFrame(
    {
        "ans": y_test,
        "pred" : y_pred.flatten()
    }
)
```

# 2교시

### 이런식으로 독립변수가 여러개 일때 PCA 로 차원줄여서 시각화할 수 있음

### 예측값과 정답값 비교 산점도

- 모델의 예측이 실제값과 얼마나 밀접하게 일치하는지 평가
- 완벽한 모델은 모든 점이 대각선 위에 놓여 있음

```python
plt.figure(figsize = (10,6))

plt.scatter(y_pred, y_test, alpha = 0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], "k--")

plt.xlabel("Pred")
plt.ylabel("Actual")
plt.show()
```

## 잔차도

- 잔차의 패턴을 식별하는 데 도움
- 이상적으로는 잔차가 0 주위에 무작위로 흩어져 있어야 하며, 이는 모델의 오류가 무작위임을 나타냄

```python
# 잔차
residuals = y_test - y_pred
```

```python
plt.figure(figsize = (10,6))

plt.scatter(y_pred, residuals, alpha = 0.5)

plt.axhline(y=0, color ="r", linestyle = "--")

plt.xlabel("Pred")
plt.ylabel("Actual")
plt.show()
```

- 잔차도에서 패턴이 잘 나타남
- 퍼지거나 동글... 이러면 좋지 못한 예측을 하고 있다라는 뜻

### QQ plot 

- 잔차 분포를 정규 분포와 비교
- 잘 맞는 모델에서는 점들이 대략 대각선을 따라 위치해야함
  

```python
plt.figure(figsize = (10,6))

stats.probplot(residuals.values.flatten(), dist = "norm", plot = plt)


plt.title("QQResiduals")
plt.show()
```

### 잔차 히스토그램

- 잔차 분포에 대한 또 다른 시각을 제공

- 잘 맞는 모델의 경우 0을 중심으로 하는 정규 분포를 볼 수 있음

```python
plt.figure()

sns.histplot(residuals, kde = True)

plt.xlabel("Pred")
plt.ylabel("Actual")
plt.show()
```

- 오른쪽 꼬리가 기니까 예측값이 좀 크지 않나~ 하는 인사이트
